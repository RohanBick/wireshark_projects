```python
!pip install scapy
import pandas as pd
from scapy.all import *

def extract_packet_features(packet):
    features = {}
    if IP in packet:
        features['SourceIP'] = packet[IP].src
        features['DestIP'] = packet[IP].dst
    if TCP in packet:
        features['SrcPort'] = packet[TCP].sport
        features['DstPort'] = packet[TCP].dport
        features['Protocol'] = 'TCP'
        features['PacketSize'] = len(packet)
    elif UDP in packet:
        features['SrcPort'] = packet[UDP].sport
        features['DstPort'] = packet[UDP].dport
        features['Protocol'] = 'UDP'
        features['PacketSize'] = len(packet)
    else:
        # If the protocol is neither TCP nor UDP, you can handle it accordingly
        features['Protocol'] = 'Other'
        features['PacketSize'] = len(packet)

    return features

def main(pcap_file, output_csv):
    packets = rdpcap(pcap_file)  # Read the PCAP file
    extracted_features = []

    for packet in packets:
        features = extract_packet_features(packet)
        if features:
            extracted_features.append(features)

    # Create a DataFrame from the extracted features
    df = pd.DataFrame(extracted_features, columns=['SourceIP', 'DestIP', 'SrcPort', 'DstPort', 'Protocol', 'PacketSize'])

    # Save the DataFrame to a CSV file
    df.to_csv(output_csv, index=False)

if __name__ == '__main__':
    pcap_file = 'iot2.pcap'  # Replace with the path to your PCAP file
    output_csv = 'output_features.csv'  # Replace with the desired output CSV file
    main(pcap_file, output_csv)
```
